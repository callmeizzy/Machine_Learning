```python
sub_df = pd.read_csv(path[2])
sub_df
```

```python
sub_df = df[['u','g','r','i','z']]
```

```python
#전치행렬
sub_df_t = np.transpose(sub_df)
```

```python
sub_df_t
```

```python
cov_df = sub_df_t.dot(sub_df)
```

```python
#e가 고윳값, v는 고유벡터  / 2개가 나오게 무조건 걸어놓음
# 고윳값 분해 : eig = eigenvalue decomposition , 유니크 
e, v = np.linalg.eig(cov_df)
e
#고윳값이 큰 순서대로 나온다 
```

```python
v
```

```python
# 특이값  //// 특이값 분해 : svd - singular value decomposition
U, S, Vt = np.linalg.svd(cov_df)
U
```

```python
S
#특이값이 큰 순서대로 나온다 
```
